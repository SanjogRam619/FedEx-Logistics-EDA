## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# ==============================================================
# Data Wrangling
# Preparing the dataset for Exploratory Data Analysis (EDA)
# ==============================================================

# Create a copy of the original dataset
# This preserves the original data for future reference.
df_clean = df.copy()

# --------------------------------------------------------------
# 1. Remove Duplicate Records
# --------------------------------------------------------------
duplicate_count = df_clean.duplicated().sum()

if duplicate_count > 0:
    df_clean.drop_duplicates(inplace=True)
    print(f"{duplicate_count} duplicate records removed.")
else:
    print("No duplicate records found.")

# --------------------------------------------------------------
# 2. Handle Missing Values
# --------------------------------------------------------------

# Fill missing Shipment Mode with the most frequent value (Mode)
df_clean['Shipment Mode'].fillna(df_clean['Shipment Mode'].mode()[0], inplace=True)

# Fill missing Line Item Insurance (USD) with Median
df_clean['Line Item Insurance (USD)'].fillna(
    df_clean['Line Item Insurance (USD)'].median(),
    inplace=True
)

# Fill missing Dosage with 'Unknown'
df_clean['Dosage'].fillna('Unknown', inplace=True)

print("Missing values handled successfully.")

# --------------------------------------------------------------
# 3. Convert Date Columns to Datetime Format
# --------------------------------------------------------------

date_columns = [
    'PQ First Sent to Client Date',
    'PO Sent to Vendor Date',
    'Scheduled Delivery Date',
    'Delivered to Client Date',
    'Delivery Recorded Date'
]

for col in date_columns:
    df_clean[col] = pd.to_datetime(df_clean[col], errors='coerce')

print("Date columns converted successfully.")

# --------------------------------------------------------------
# 4. Verify Dataset After Cleaning
# --------------------------------------------------------------

print("\nRemaining Missing Values:")
display(df_clean.isnull().sum()[df_clean.isnull().sum() > 0])

print("\nDataset Shape After Data Wrangling:", df_clean.shape)
````

### What all manipulations have you done and insights you found?



During the data wrangling phase, the dataset was prepared for analysis by improving its overall quality and consistency. A copy of the original dataset was created to preserve the raw data. The dataset was checked for duplicate records, and no duplicate entries were found, indicating good data integrity. Missing values were handled appropriately by replacing missing values in the Shipment Mode column with the most frequent category (mode), filling missing values in Line Item Insurance (USD) with the median value to minimize the impact of outliers, and replacing missing values in the Dosage column with "Unknown" to retain all records. Additionally, all delivery-related date columns were converted into the datetime format to enable accurate time-based analysis. After these preprocessing steps, the dataset became clean, consistent, and ready for exploratory data analysis. These transformations ensure reliable statistical analysis and improve the quality of the insights generated from the dataset.

````python
# ==============================================================
# Chart 5: Distribution of Freight Cost (USD)
# ==============================================================

freight = df['Freight Cost (USD)'].dropna()

upper_limit = freight.quantile(0.99)

freight = freight[freight <= upper_limit]

plt.figure(figsize=(10,6))

sns.histplot(
    freight,
    bins=40,
    kde=True,
    color='royalblue',
    edgecolor='black'
)

plt.title("Distribution of Freight Cost (USD)", fontsize=16, fontweight='bold')
plt.xlabel("Freight Cost (USD)")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.4)

plt.tight_layout()
plt.show()
````

#### Chart - 13

````python
# ==============================================================
# Chart 13: Top 10 Countries vs Shipment Mode vs Average Freight Cost
# ==============================================================

# Remove missing freight cost values
df_chart13 = df.dropna(subset=['Freight Cost (USD)'])

# Select top 10 countries by shipment count
top10_countries = df['Country'].value_counts().head(10).index

# Filter data
chart13 = df_chart13[df_chart13['Country'].isin(top10_countries)]

# Calculate average freight cost
chart13 = (
    chart13.groupby(['Country', 'Shipment Mode'])['Freight Cost (USD)']
    .mean()
    .reset_index()
)

plt.figure(figsize=(14,7))

ax = sns.barplot(
    data=chart13,
    x='Country',
    y='Freight Cost (USD)',
    hue='Shipment Mode',
    palette='viridis'
)

# Format Y-axis in Thousands (K)
from matplotlib.ticker import FuncFormatter

ax.yaxis.set_major_formatter(
    FuncFormatter(lambda x, pos: f'{x/1000:.0f}K')
)

plt.title(
    "Average Freight Cost by Country and Shipment Mode",
    fontsize=16,
    fontweight='bold'
)

plt.xlabel("Country", fontsize=12)
plt.ylabel("Average Freight Cost (USD)", fontsize=12)

plt.xticks(rotation=30, ha='right')

plt.grid(axis='y', linestyle='--', alpha=0.4)

plt.legend(title="Shipment Mode")

plt.tight_layout()
plt.show()
````

A grouped bar chart is ideal for comparing the average freight cost across multiple countries while simultaneously distinguishing between different shipment modes. It enables clear comparison of logistics costs across geographical regions and transportation methods in a single visualization.

##### 2. What is/are the insight(s) found from the chart?

The chart shows that the average freight cost varies considerably across both countries and shipment modes, indicating that transportation expenses depend on a combination of destination and shipping method. Ocean shipments generally have the highest average freight cost in several countries, including Côte d'Ivoire, Mozambique, Tanzania, and Zambia, suggesting they are relatively expensive for these destinations. Air Charter records the highest average freight cost in countries such as Nigeria and Zimbabwe, reflecting the premium cost of charter services for specific routes. In contrast, Truck shipments generally have lower average freight costs across most countries, making them a more economical transportation option where applicable. These differences highlight that logistics costs are influenced by both geographical location and the chosen shipment mode, emphasizing the need for route-specific transportation planning.

#### Chart - 14 - Correlation Heatmap

````python
# ==============================================================
# Chart 14: Correlation Heatmap
# ==============================================================

# Select only numerical columns
numeric_df = df.select_dtypes(include=['int64', 'float64'])

# Calculate correlation matrix
correlation_matrix = numeric_df.corr()

# Plot heatmap
plt.figure(figsize=(12,8))

sns.heatmap(
    correlation_matrix,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    linewidths=0.5,
    square=True
)

plt.title(
    "Correlation Heatmap of Numerical Variables",
    fontsize=16,
    fontweight="bold"
)

plt.tight_layout()
plt.show()
````

##### 1. Why did you pick the specific chart?

A correlation heatmap is an ideal multivariate visualization because it simultaneously displays the relationships among multiple numerical variables. It helps identify strong positive and negative correlations, detect redundant variables, and understand how different business metrics influence one another.

##### 2. What is/are the insight(s) found from the chart?

The correlation heatmap reveals several important relationships among the numerical variables in the dataset. The strongest positive correlation is between Line Item Value and Line Item Insurance (USD) (0.96), indicating that higher-value shipments generally require higher insurance costs. There is also a strong positive correlation between Line Item Quantity and Line Item Value (0.84), suggesting that shipments with more items tend to generate greater business value. Additionally, Line Item Quantity and Line Item Insurance (USD) show a high positive correlation (0.80), indicating that larger shipments usually incur higher insurance costs.

Freight Cost (USD) has a moderate positive correlation with both Line Item Value (0.43) and Line Item Quantity (0.38), implying that transportation costs tend to increase as shipment size and value increase, though other factors also influence freight charges. Most remaining variable pairs have weak or negligible correlations, suggesting they are relatively independent and do not strongly influence each other.
